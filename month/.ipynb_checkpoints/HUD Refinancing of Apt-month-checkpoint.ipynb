```python
cities=["BOSTON", "PHILADELPHIA", "NEW YORK", "WASHINGTON", "CHICAGO", "LOS ANGELES", "ATLANTA"]
```

```python
nc=pd.read_csv("City Data of HUD_Refinancing of Apt_month_selected city.csv")
```

```python
def to_month(cc):
    cc["Month"]=cc["Month"].map(lambda x:pd.to_datetime(x,format='%Y-%m'))
    cc["Month"]=cc["Month"].map(lambda x:x.to_period("M"))
    return cc
```

```python
def add_corr(cc):
    
    cc["Corr with mortgage 30_89"]=cc["INTEREST RATE"].corr(cc["30-89"])
    cc["Corr with lag 1 mortgage 30_89"]=cc["INTEREST RATE"].corr(cc["30-89(-1)"])
    cc["Corr with lag 2 mortgage 30_89"]=cc["INTEREST RATE"].corr(cc["30-89(-2)"])
    cc["Corr with lag 3 mortgage 30_89"]=cc["INTEREST RATE"].corr(cc["30-89(-3)"])
    cc["Corr with lag 4 mortgage 30_89"]=cc["INTEREST RATE"].corr(cc["30-89(-4)"])
    cc["Corr with lead 1 mortgage 30_89"]=cc["INTEREST RATE"].corr(cc["30-89(1)"])
    cc["Corr with lead 2 mortgage 30_89"]=cc["INTEREST RATE"].corr(cc["30-89(2)"])
    cc["Corr with lead 3 mortgage 30_89"]=cc["INTEREST RATE"].corr(cc["30-89(3)"])
    cc["Corr with lead 4 mortgage 30_89"]=cc["INTEREST RATE"].corr(cc["30-89(4)"])
    cc["Corr with mortgage 90"]=cc["INTEREST RATE"].corr(cc["90"])
    cc["Corr with lag 1 mortgage 90"]=cc["INTEREST RATE"].corr(cc["90(-1)"])
    cc["Corr with lag 2 mortgage 90"]=cc["INTEREST RATE"].corr(cc["90(-2)"])
    cc["Corr with lag 3 mortgage 90"]=cc["INTEREST RATE"].corr(cc["90(-3)"])
    cc["Corr with lag 4 mortgage 90"]=cc["INTEREST RATE"].corr(cc["90(-4)"])
    cc["Corr with lead 1 mortgage 90"]=cc["INTEREST RATE"].corr(cc["90(1)"])
    cc["Corr with lead 2 mortgage 90"]=cc["INTEREST RATE"].corr(cc["90(2)"])
    cc["Corr with lead 3 mortgage 90"]=cc["INTEREST RATE"].corr(cc["90(3)"])
    cc["Corr with lead 4 mortgage 90"]=cc["INTEREST RATE"].corr(cc["90(4)"])
    return cc
```

```python
for i in range(len(cities)):
    sub=nc[nc["PROPERTY CITY"]==cities[i]]
    if len(sub)>10:
        sub=to_month(sub)
        mn=cities[i]+" mortgage.csv"
        subm=pd.read_csv(mn)
        subm=subm.rename(columns={"Unnamed: 0":"Month"})
        subm=to_month(subm)
        sub=sub.merge(subm,on="Month")
        sub=add_corr(sub)
        sub.to_csv("ref of apt/"+cities[i]+" all.csv")
        
```

```python
def add_mort(cc):
    cc=to_month(cc)
    name=cc["PROPERTY CITY"].iloc[0]
    mn=name+" mortgage.csv"
    ccm=pd.read_csv(mn)
    ccm=ccm.rename(columns={"Unnamed: 0":"Month"})
    ccm=to_month(ccm)
    cc=cc.merge(ccm,on="Month")
    cc=add_corr(cc)
    return cc
    
```

```python
at=nc[nc["PROPERTY CITY"]=="ATLANTA"]
at=add_mort(at)
at_sub=at[["PROPERTY CITY","Month","INTEREST RATE","auto_vol(3)","student_vol","credit_tightness(-2)",
                  "30-89(4)","90(4)"]]
at_sub.to_csv("loan/refin apt_ATLANTA.csv")
```

```python
for i in range(len(cities)):
    sub=nc[nc["PROPERTY CITY"]==cities[i]]
    if (len(sub))>=10:
        x=sub["Month"].map(lambda x:str.split(x,"/")[-1])
        plt.figure(figsize=(20,10))
        plt.grid(dash)
        plt.plot(x, sub["INTEREST RATE"], label="Loan Rate",color="black")
        plt.plot(x , sub["ON"],label="ON Libor")
        plt.plot(x , sub["1W"],label="1 Week Libor")
        plt.plot(x , sub["1M"],label="1 Month Libor")
        plt.plot(x , sub["6M"],label="6 Month Libor")
        plt.plot(x, sub["12M"], label="12 Month Libor")
        plt.plot(x , sub["1 Yr"],label="1 Yr Treasury rate")
        plt.plot(x , sub["5 Yr"],label="5 Yr Treasury rate")
        plt.plot(x , sub["10 Yr"],label="10 Yr Treasury rate")
        plt.plot(x , sub["20 Yr"],label="20 Yr Treasury rate")
        plt.plot(x, sub["30 Yr"], label="30 Yr Treasury rate")
        plt.ylabel('rate %',fontsize = 16)
        plt.title(cities[i]+"_Refinancing of Apt_Loan Rate with Different Rate")
        plt.xticks(np.arange(len(sub), step=len(sub)/12))
        plt.legend(loc="upper right")
        plt.savefig('figure/'+cities[i]+'_Refinancing of Apt_Loan Rate with Different Rate_month.jpg')
```
