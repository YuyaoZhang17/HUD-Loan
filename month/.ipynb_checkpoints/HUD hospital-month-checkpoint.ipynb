```python
import pandas as pd
import numpy as np
import datetime
import glob
import plotly
import dash
from matplotlib import pyplot as plt
```

```python
start="2009-10-14"
```

```python
libor=pd.read_csv("../LIBOR USD.csv")
```

```python
libor["Date"]=libor["Date"].map(lambda x:pd.to_datetime(x,format='%d.%m.%Y'))
```

```python
libor["Month"]=libor["Date"].map(lambda x:x.to_period("M"))
```

```python
libor=libor.sort_values("Date")
```

```python
libor=libor[libor["Date"]>start]
libor=libor.groupby("Month")["ON","1W","1M","2M","3M","6M","12M"].mean()
```

```python
path_trea =r'../Treasury' 
allFiles_trea = glob.glob(path_trea + "/*.csv")
frame_trea = pd.DataFrame()
list_trea = []
for file_trea in allFiles_trea:
    df_trea = pd.read_csv(file_trea,index_col="Date", header=0)
    list_trea.append(df_trea)
frame_trea=pd.concat(list_trea)
```

```python
frame_trea.index=frame_trea.index.map(lambda x:pd.to_datetime(x,format='%m/%d/%Y'))
```

```python
frame_trea["Month"]=frame_trea.index.map(lambda x:x.to_period("M"))
```

```python
frame_trea.to_csv("Treasury_month.csv")
```

```python
frame_trea=frame_trea[frame_trea.index>start]
frame_trea=frame_trea.groupby("Month")["1 Mo","6 Mo","1 Yr","5 Yr","10 Yr","20 Yr","30 Yr"].mean()
```

```python
health=pd.read_csv("City Data of HUD_Hospitals and Group Practices_month.csv")
```

```python
cc=health[health["PROPERTY CITY"]=="CHICAGO"]
print(len(cc))
```

```python
x=cc["Month"].map(lambda x:str.split(x,"/")[-1])
plt.figure(figsize=(20,10))
plt.grid(dash)
plt.plot(x, cc["INTEREST RATE"], label="Loan Rate",color="black")
plt.plot(x , cc["ON"],label="ON Libor")
plt.plot(x , cc["1W"],label="1 Week Libor")
plt.plot(x , cc["1M"],label="1 Month Libor")
plt.plot(x , cc["6M"],label="6 Month Libor")
plt.plot(x, cc["12M"], label="12 Month Libor")
plt.plot(x , cc["1 Yr"],label="1 Yr Treasury rate")
plt.plot(x , cc["5 Yr"],label="5 Yr Treasury rate")
plt.plot(x , cc["10 Yr"],label="10 Yr Treasury rate")
plt.plot(x , cc["20 Yr"],label="20 Yr Treasury rate")
plt.plot(x, cc["30 Yr"], label="30 Yr Treasury rate")
plt.ylabel('rate %',fontsize = 16)
plt.title("Boston_Hospital_Loan Rate with Different Rate")
plt.xticks(np.arange(len(cc), step=5))
plt.legend(loc="upper right")
plt.savefig('figure/Boston_Hospital_Loan Rate with Different Rate_month.jpg')
```
